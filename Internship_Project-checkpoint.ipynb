1.Problem Statement

a new Fast Moving Consumer Goods (fmcg) goods company entered the food market scene with instant noodles as their product.they are facing a problem of
supply chain and inventory not working as intended.this causes a supply and demand problem,in places where there is a high demand for their products the 
inventory of the said product is very low and vice-versa.this will cause huge loses for the company as they will lose their customers due to lack of products and also they are wasting inventory cost where product demand is low.the management want to fix this problem by optimizing the supply chain and
inventory as needed.

Objective

To provide an understanding as to why the inventory is not working as it is supposed to and to help optimize inventory levels, ensuring that high-demand locations have sufficient stock while minimizing excess inventory in low-demand locations using machine learning techniques.
